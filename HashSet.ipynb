# Hash Set 
In this notebook I validate unique hashes for the datasets.

```python
import mt_util as mt
import pandas as pd
import math
import hashlib
import shutil
# Helpful Functions
def getHWdim(size):
    x = int(math.sqrt(size))
    return x 
def getFileType(x):
    ss = x.split('.')
    return ss[len(ss) - 2]
def hashFile(path):
    return hashlib.md5(open(path,'rb').read()).hexdigest()
# Importing Image class from PIL module 
from PIL import Image 
import pathlib
def resizeIM(im_path, save_path, file_name):
    try:
        # Opens a image in RGB mode
        im = Image.open(im_path)
        newsize = (150, 150)
        im1 = im.resize(newsize)
        file_name = str(file_name) + '.png'
        save_path = save_path + "\\" + pathlib.PureWindowsPath(file_name).name
        im1.save(save_path)
    except Exception as e:
        print(e)
```

## Validate Exe Data

```python
#data folder contains generated exe files
path = "data"
# Create the pandas DataFrame from object dictionary
df = pd.DataFrame([o.__dict__ for o in mt.get_files(path)])

# print dataframe.
df.head(5)
```

```python
df
```

```python
df['dim'] = df['size'].apply(lambda x: getHWdim(x))
df['md5'] = df['path'].apply(lambda x: hashFile(x))
df
```

```python
df.md5.value_counts().mean()
```

```python
mf = df.md5.value_counts().to_frame()
mf
```

```python
mf.query("md5 == 1")
```

```python
df.to_csv("combined_exes.csv")
mf.to_csv("combined_exes_hashes.csv")
```
